```python
# to extract text from html_document
def extract_text(html_document):
        from bs4 import BeautifulSoup
        article_soup = BeautifulSoup(html_document, 'lxml')
        text = article_soup.find_all(text=True)
        
        black_list = [
            'script',
            'noscript',
            '[document]',
            'header',
            'html',
            'meta',
            'head',
            'input'
        ]
        
        output = ""
        for t in text:
            #print("t:", t)
            if t.parent.name not in black_list:
                output += t
            #print("output:", output)
        output = output.replace('\n', '').strip().replace(' ', '')
        
        return output
    
# html_document = '<div>\n<div style="text-align:center;"><img src="http://taiwan101.itaiwan.net//uploaded_images/1_0.jpg?1500403865" alt="1_0.jpg?1500403865" /></div>\n\n<div>dlfksjkdflfldsf</div>\n\n<div>sskdfslfjslf</div>\n\n<div>\n<div style="text-align:center;"><img src="http://taiwan101.itaiwan.net//uploaded_images/1_1.jpg?1500403888" alt="1_1.jpg?1500403888" /></div>\n\n<div>\xa0</div>\n\n<div style="text-align:center;"><iframe allowfullscreen="" frameborder="0" src="https://player.vimeo.com/video/223518754?ecver=1"></iframe></div>\n\n<div style="text-align:center;">\xa0</div>\n\n<div style="text-align:center;">http://www.ntu.edu.tw/</div>\n\n<div style="text-align:center;"><iframe allowfullscreen="" frameborder="0" src="https://www.youtube.com/embed/-sw8yhZaYes?ecver=1"></iframe></div>\n\n<div style="text-align:center;">\xa0</div>\n\n<div style="text-align:center;"><img src="http://www.taiwanlikes.com/uploaded_images/1_2.jpg?1503993988" alt="1_2.jpg?1503993988" /></div>\n</div>\n</div>\n'
# #html_document = '<div><img src="http://www.taiwanlikes.com/uploaded_images/113_0.jpg?1503247969" alt="113_0.jpg?1503247969" /></div>\n\n<div>\xa0</div>\n\n<div>\xa0</div>\n\n<div>原標題：詩詞日曆｜低頭弄蓮子，蓮子清如水 西洲曲 南朝 樂府民歌 憶梅下西洲，折梅寄江北。 單衫杏子紅，雙鬢鴉雛色。 西洲在何處？兩槳橋頭渡。 日暮伯勞飛，風吹烏臼樹。 樹下即門前，門中露翠鈿。 開門郎不至，出門采紅蓮。 採蓮南塘秋，蓮花過人頭。 低頭弄蓮子，蓮子清如水。</div>\n\n<div>\xa0</div>\n\n<div><img src="https://www.iread.one/images/image/44/8805533.jpeg?t=1503187084" alt="8805533.jpeg?t=1503187084" /></div>\n\n<div>\xa0</div>\n\n<div>置蓮懷袖中，蓮心徹底紅。 憶郎郎不至，仰首望飛鴻。 鴻飛滿西洲，望郎上青樓。 樓高望不見，盡日欄桿頭。 欄杆十二曲，垂手明如玉。 捲簾天自高，海水搖空綠。 海水夢悠悠，君愁我亦愁。 南風知我意，吹夢到西洲。 鴉雛色：像小烏鴉一樣的顏色。形容女子的頭髮烏黑髮亮。 伯勞：鳥名。 烏臼（jiù）：就是烏桕（jiù）樹。 翠鈿（diàn）：用翠玉做成或鑲嵌的首飾。 望飛鴻：望飛雁，希望鴻雁傳書。 【大詩兄說】我們今天所說的南京城，在南北朝時期的宋齊梁陳這四個南方朝廷，叫做建康，這是他們的都城。出建康城的西門，長江江心之中有一塊沖積沙洲，當時的人們都叫它西洲。長江泥沙富含營養，西洲成了一塊沃土，是動植物的樂園。 早春時節，紅梅綻放，漁家女搖船登洲，折下一枝梅花，在把船兒搖到對岸的江北——沒錯，現在叫做浦口——送給住在這兒的心上人。 夏天的時候，沙洲上一大片青紗帳、蘆葦盪，隨風搖擺。伯勞鳥兒飛過，烏桕樹綠葉婆娑。西洲上有幾處淺灘，不知道哪一年哪一月，有人丟棄了幾顆吃剩的蓮蓬，居然就生長出一片茂盛的蓮荷。 夏末秋初，姑娘搖船來到西洲，採集蓮蓬。剝出青青的蓮子，凝視著出神。這蓮子純凈無暇，就像這青草沙中的清水。 秋色漸濃，鴻雁從北方飛來，把一行行影子投在江面上。姑娘仰視飛鴻。聽說鴻雁是郵差，會幫人帶信。不過，姑娘從來沒有等到過一封書信。其實，自從送了梅花之後，她再也沒有見過他。 這首民歌好長啊！它就得這麼長，否則怎麼能把西洲姑娘的「四季歌」唱完呢？ <strong>今天（8月20日）下午1點到2點，上海書城福州路店三樓，舉行《一年好景君須記：古典詩詞中的季節之美》新書講座暨簽售會。</strong> <strong>大詩兄與你相約，不見不散！</strong></div>\n\n<div>\xa0</div>\n\n<div><img src="https://www.iread.one/images/image/44/8805543.jpeg?t=1503187086" alt="8805543.jpeg?t=1503187086" /></div>\n'
# text = extract_text(html_document)
# print(text)
```

```python
# to include only Chinese (excluding punctuation, number, and alphabet)
def is_chinese(uchar):
    if uchar >= u'\u4E00' and uchar <= u'\u9fa5':
        return True
    else:
        return False
# def is_punctuation(uchar):
#     if ()

def format_string(content):
    content_str = ""
    for i in content:
        if is_chinese(i):
            content_str += i
        else:
            content_str += " " # to add space when the characteristic is not Chinese
    content_str = content_str.strip()
    content_str = content_str.replace("  ", " ")
    return content_str
line = """dlfksjkdflfldsfsskdfslfjslf  http://www.ntu.edu.tw/

汪洋潛水降龍十八掌之一，潛水3大基本觀念。 潛水生涯中，許多觀念，都很重要。 因為潛水，最基本上，是一種「觀察、思考、判斷、行動」的「腦內運動」。 所以，許多潛水的觀念、體會、與經驗，相當重要。 在此先簡單聊一下，潛水裡面比較基本的3大觀念。 1、不可憋氣。 因為你一憋氣，或是沒有把氣吐盡，你就無法順利排氮，排出人體無法消化的氮氣與廢氣。 2、須與潛伴保持在3-5公尺內。 不可距離潛伴過遠。 3、上升的速度，越慢越好，每上升1公尺，約花5-10秒。 上升的速度，不可比你的氣泡還快。"""
result = format_string(line)
print(result)
```

```python
# to read the article as corpus from DATABASE: contentanalysis TABLE: articles
def get_corpus_from_mysql(sql_database, sql_table, column_name):
    article_text = ""
    
    import mysql.connector
    sql_db = mysql.connector.connect(
        host = '127.0.0.1',
        user = 'root',
        password = '',
        database = sql_database
    )
    
    cursor = sql_db.cursor()
    
    # to check whether TABLE exists or not
    cursor.execute("SHOW TABLES LIKE '{}'".format(sql_table))
    temp_result = cursor.fetchone()
    if (temp_result):
        logging.info("TABLE {} exists in DATABASE {}".format(sql_table, sql_database))
    else:
        logging.info("TABLE {} exists in DATABASE {}".format(sql_table, sql_database))
        
    # to select the value in column_name
    select_sql = "SELECT {} FROM {}".format(column_name, sql_table)
    logging.info("SELECT syntax: {}".format(select_sql))
    cursor.execute(select_sql)
    result = cursor.fetchall()
    logging.info("the amount of the content retrieve from DATABASE {}: {}".format(sql_database, str(len(result))))
    # to output the raw text extracted as a .txt file
    output_filename = "advertisement_article_raw_text_0617.txt"
    output_f = open(output_filename, 'w', encoding='utf-8')
    text_num = 0
    for text in result:
        temp_t = extract_text(text[0]) # to extract text(tuple's value) from html_document
        temp_t = format_string(temp_t) # to include only Chinese (excluding punctuation, number, and alphabet)
        output_f.write(temp_t+'\n')
        text_num += 1
        if text_num %1000 == 0:
            logging.info("Have processed: {} articles".format(str(text_num)))
        article_text = article_text+'\n'+temp_t

    return article_text, output_filename


sql_database = "contentanalysis"
sql_table = "articles"
column_name = "content"
article_text, output_filename = get_corpus_from_mysql(sql_database, sql_table, column_name)
#result = get_corpus_from_mysql(sql_database, sql_table, column_name)
logging.info("Have extract text from the content html from TABLE {} and output as txt file {}".format(sql_table, output_filename))
print(article_text[0:10000])
```

```python
# to build up a word2vec model using gensim.model package
from gensim.models import word2vec

source = "advertisement_article_cut.txt"
model_name = "ad_w2v_model_v1"
vector_size = 100
min_count = 1
iter_count = 10
window_size = 10
workers = 3
sg_mode = 1 # 1=skip-gram; 0=CBOW

sentences = word2vec.LineSentence(source)
w2v_model = word2vec.Word2Vec(sentences, size=vector_size, iter=iter_count, sg=sg_mode, min_count=min_count, window=window_size, workers=workers)
w2v_model.save(model_name)
logging.info("Trained & Saved {}".format(model_name))
```

```python
# to load the word2vec model
from gensim import models
model_name = "ad_w2v_model_v1"
w2v_model = models.Word2Vec.load(model_name)
logging.info("inied loading w2v model: {}".format(model_name))
```
