## 1. Load Data

```python
import logging
logging.basicConfig(format='%(asctime)s: %(levelname)s: %(message)s', level=logging.INFO)
```

```python
import pandas as pd
ad_data = pd.read_csv('C:/Users/ching870423/extract_text_practice/ad_feature_with_img.csv')
print(len(ad_data))
print(ad_data.columns)
ad_data.head(10)
```

```python
ad_data_le = ad_data[['WEEKDAY', 'HOUR', 'CATEGORY', 'fans',
       'ent0', 'ent1', 'ent2', 'ent3', 'ent4', 'ent5', 'ent6', 'ent7', 'ent8',
       'ent9', 'ent10', 'ent11', 'ent12', 'ent13', 'ent14', 'ent15', 'ent16',
       'ent17', 'ent18', 'ent19', 'ent20', 'ent21', 'ent22', 'ent23', 'ent24',
       'ent25', 'ent26', 'ent27', 'ent28', 'img_width_pixel',
       'img_height_pixel', 'img_red', 'img_green', 'img_blue']]

from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
ad_data_le['CATEGORY_le'] = labelencoder.fit_transform(ad_data_le['CATEGORY'])
ad_data_cat = ad_data_le[['CATEGORY', 'CATEGORY_le']]
ad_data_cat
```

```python
# to transform the label encoding to one-hot-encoding
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(handle_unknown='ignore')
CAT_enc_df = pd.DataFrame(enc.fit_transform(ad_data_cat[['CATEGORY_le']]).toarray())
ad_test_df = ad_data_cat.join(CAT_enc_df)
# to rename the columns
cols_name = ['CATEGORY', 'CATEGORY_le']
for cat in labelencoder.classes_:
    cols_name.append(cat)
ad_test_df.columns = cols_name
#ad_test_df.head(10)

# one-hot-encoding WEEKDAY
WEEK_enc_df = pd.DataFrame(enc.fit_transform(ad_data_le[['WEEKDAY']]).toarray())
ad_test_WEEK_df = pd.DataFrame(ad_data_le['WEEKDAY']).join(WEEK_enc_df)
# to rename the columns
cols_name_WEEK = ['WEEKDAY']
for i in range(7):
    temp_name = 'WEEKDAY_{}'.format(str(i))
    cols_name_WEEK.append(temp_name)
ad_test_WEEK_df.columns = cols_name_WEEK
#ad_test_WEEK_df

# one-hot-encoding HOUR
HOUR_enc_df = pd.DataFrame(enc.fit_transform(ad_data_le[['HOUR']]).toarray())
ad_test_HOUR_df = pd.DataFrame(ad_data_le['HOUR']).join(HOUR_enc_df)
# to rename the columns
cols_name_HOUR = ['HOUR']
for i in range(24):
    temp_name = 'HOUR_{}'.format(str(i))
    cols_name_HOUR.append(temp_name)
ad_test_HOUR_df.columns = cols_name_HOUR
ad_test_HOUR_df
```

```python
# drop the exist column in total DataFrame
ad_data_enc = ad_data_le.drop(['CATEGORY', 'CATEGORY_le', 'WEEKDAY', 'HOUR'], axis=1)

# process CATEGORY
ad_test_df = ad_test_df.drop(['CATEGORY', 'CATEGORY_le'], axis=1)
ad_data_enc = ad_data_enc.join(ad_test_df)

# process WEEKDAY
ad_test_WEEK_df = ad_test_WEEK_df.drop(['WEEKDAY'], axis=1)
ad_data_enc = ad_data_enc.join(ad_test_WEEK_df)

# process HOUR
ad_test_HOUR_df = ad_test_HOUR_df.drop(['HOUR'], axis=1)
ad_data_enc = ad_data_enc.join(ad_test_HOUR_df)

ad_data_enc.columns
```

```python
ad_data_enc['label90'] = ad_data['label90']
# to create area for img
ad_data_enc['img_area_pixel'] = ad_data_enc['img_width_pixel']*ad_data_enc['img_height_pixel']
print(ad_data_enc.columns)
ad_data_enc.head(10)
```

```python
x_enc_cols = ['fans', 'ent0', 'ent1', 'ent2', 'ent3', 'ent4', 'ent5', 'ent6', 'ent7',
       'ent8', 'ent9', 'ent10', 'ent11', 'ent12', 'ent13', 'ent14', 'ent15',
       'ent16', 'ent17', 'ent18', 'ent19', 'ent20', 'ent21', 'ent22', 'ent23',
       'ent24', 'ent25', 'ent26', 'ent27', 'ent28', 'img_width_pixel',
       'img_height_pixel', 'img_red', 'img_green', 'img_blue', 'Adult',
       'Arts & Entertainment', 'Autos & Vehicles', 'Beauty & Fitness',
       'Books & Literature', 'Business & Industrial',
       'Computers & Electronics', 'Finance', 'Food & Drink', 'Games', 'Health',
       'Hobbies & Leisure', 'Home & Garden', 'Internet & Telecom',
       'Jobs & Education', 'Law & Government', 'News', 'Online Communities',
       'People & Society', 'Pets & Animals', 'Real Estate', 'Reference',
       'Science', 'Sensitive Subjects', 'Shopping', 'Sports', 'Travel',
       'WEEKDAY_0', 'WEEKDAY_1', 'WEEKDAY_2', 'WEEKDAY_3', 'WEEKDAY_4',
       'WEEKDAY_5', 'WEEKDAY_6', 'HOUR_0', 'HOUR_1', 'HOUR_2', 'HOUR_3',
       'HOUR_4', 'HOUR_5', 'HOUR_6', 'HOUR_7', 'HOUR_8', 'HOUR_9', 'HOUR_10',
       'HOUR_11', 'HOUR_12', 'HOUR_13', 'HOUR_14', 'HOUR_15', 'HOUR_16',
       'HOUR_17', 'HOUR_18', 'HOUR_19', 'HOUR_20', 'HOUR_21', 'HOUR_22',
       'HOUR_23', 'img_area_pixel'] # with img_area_pixel
y_enc_cols = ['label90']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
ad_data_enc[x_enc_cols], ad_data_enc[y_enc_cols], test_size=0.3, random_state=0)

# to check the size of the train & test
print("Train data length: {}".format(str(len(X_train))))
print("Test data length: {}".format(str(len(X_test))))
```

```python
print("Original dataset shape: \n{}".format(y_train['label90'].value_counts()))
# to oversample the ad_data_enc by SMOTE
from imblearn.over_sampling import SMOTE
sm = SMOTE(sampling_strategy=0.4, random_state=0)
X_train_rus, y_train_rus = sm.fit_resample(X_train, y_train)
print("Resampled dataset shape: \n{}".format(y_train_rus['label90'].value_counts()))
```

## Build up Keras DNN model
choose the nodes within hidden layer

```python
# import libraries
from keras.models import Sequential
from keras.layers import Dense

# find the best number of nodes in hidden layer: 48-64
best_node_num = 0
best_acc = 0.0
for i in range(48, 65):
    # define Keras model
    logging.info("Start hidden nodes: {}".format(i))
    model = Sequential()
    #model.add(Dense(i, input_dim=len(X_train.columns), activation='relu'))
    model.add(Dense(i, input_dim=len(X_train_rus.columns), activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    
    # complie the keras model
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    
    # fit the keras model on the dataset
    epoch_size = 20
    batch_size = 100
    #model.fit(X_train, y_train, epochs=epoch_size, batch_size=batch_size)
    model.fit(X_train_rus, y_train_rus, epochs=epoch_size, batch_size=batch_size)
    
    # evaluate Keras Model
    #loss, accuracy = model.evaluate(X_train, y_train)
    loss, accuracy = model.evaluate(X_train_rus, y_train_rus)
    print('model evalutaion accuracy: %.2f' % (accuracy*100))

    # make predictions
    predictions = model.predict(X_test)
    y_pred = [round(x[0]) for x in predictions]
    
    # calculate prediction accuracy
    from sklearn.metrics import accuracy_score
    pre_acc = accuracy_score(y_test, y_pred)
    print("accuracy: {:.5}".format(pre_acc))
    
    if pre_acc>best_acc:
        best_node_num = i
        best_acc = pre_acc
    
    logging.info("nodes {}, prediction accuracy {} DONE".format(i, pre_acc))
    
logging.info("Finish finding best hidden nodes number: best_node_num: {}, best_acc: {}".format(best_node_num, best_acc))


# build a DNN model with the number of nodes in hidden layer we just find
logging.info("Using best_node_num to get the final figure")
model = Sequential()
#model.add(Dense(best_node_num, input_dim=len(X_train.columns), activation='relu'))
model.add(Dense(best_node_num, input_dim=len(X_train_rus.columns), activation='relu'))
model.add(Dense(1, activation='sigmoid'))
    
# complie the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    
# fit the keras model on the dataset
epoch_size = 20
batch_size = 100
#model.fit(X_train, y_train, epochs=epoch_size, batch_size=batch_size)
model.fit(X_train_rus, y_train_rus, epochs=epoch_size, batch_size=batch_size)
    
# evaluate Keras Model
#loss, accuracy = model.evaluate(X_train, y_train)
loss, accuracy = model.evaluate(X_train_rus, y_train_rus)
print('model evalutaion accuracy: %.2f' % (accuracy*100))

# make predictions
predictions = model.predict(X_test)
y_pred = [round(x[0]) for x in predictions]

# to print out the result: accuracy, precision, recall, f1csore
# to calculate the precision
from sklearn.metrics import accuracy_score
print("accuracy: {:.5}".format(accuracy_score(y_test, y_pred)))
from sklearn.metrics import precision_score
print("precision: {:.5}".format(precision_score(y_test, y_pred, average='macro')))
from sklearn.metrics import recall_score
print("recall: {:.5}".format(recall_score(y_test, y_pred)))
from sklearn.metrics import f1_score
print("f1 score: {:.5}".format(f1_score(y_test, y_pred, average='macro')))

# to print out confusion matrix
from sklearn.metrics import confusion_matrix
tn, fp, fn, tp = confusion_matrix(y_true=y_test, y_pred=y_pred).ravel()
matrix_cols = ['Classified as Normal', 'Classified as H.P.']
matrix_index = ['Normal', 'H.P.']
matrix_result = {'Classified as Normal':[tn, fn],
                'Classified as H.P.':[fp, tp]}
confusion_matrix_df = pd.DataFrame(matrix_result, index=matrix_index, columns=matrix_cols)
confusion_matrix_df
```
